```python
import pandas as pd
import yfinance as yf 
import matplotlib.pyplot as plt
```

```python
yf.Ticker('AAPL')
```

```python
dict(yf.Ticker('AAPL').fast_info)
```

```python
tickers = ['AAPL', 'MSFT', 'IBM']
```

```python
infos = []

for i in tickers:
  infos.append(dict(yf.Ticker(i).fast_info))
```

```python
infos
```

```python
infos[0] = dict(infos[0], Ticker = "AAPL")
infos[1] = dict(infos[1], Ticker = "MSFT")
infos[2] = dict(infos[2], Ticker = "IBM")
```

```python
fundamentals = ['previousClose', 'yearHigh', 'yearLow', 'marketCap']
```

```python
df = pd.DataFrame(infos)
```

```python
df
```

```python
df = df.set_index('Ticker')
```

```python
df
```

```python
df[df.columns[df.columns.isin(fundamentals)]]
```

```python
plt.bar(df.index, df.marketCap, color = ('k', 'orange', 'cyan'))
plt.show
```

```python
plt.bar(df.index, df.yearHigh, color = ('k', 'orange', 'cyan'))
plt.show
```

```python
plt.bar(df.index, df.yearLow, color = ('k', 'orange', 'cyan'))
plt.show
```

```python
tickers = df2.Symbol.to_list()
```

```python
tickers
```

```python
infos = []

for i in tickers:
  infos.append(dict(yf.Ticker(i).fast_info))
```

```python
infos
```

```python
df3 = df3.set_index('Ticker')
```

```python
df3
```

```python
df3.shares.nlargest(5)
```

```python
df3.marketCap.nlargest(5)
```

```python
df3.shares.nsmallest(5)
```

```python
df3.lastPrice.nlargest(5)
```
